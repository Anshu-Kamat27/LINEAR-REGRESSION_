```python
def LinearRegression(w,b,x):
    return w*x+b
```

```python
def Squared_Loss_function(y,y1):
    return ((y-y1)**2)
```

```python
def Gradient_Cost_function(Inputset, Labelset, size, weight, bias):
    GradientCostForWeights = 0
    GradientCostForBias = 0

    for i in range(size):
        GradientCostForWeights += Inputset[i] * (
            LinearRegression(weight, bias, Inputset[i]) - Labelset[i]
        )
        GradientCostForBias += (
            LinearRegression(weight, bias, Inputset[i]) - Labelset[i]
        )

    return [
        (2 * GradientCostForWeights) / size,
        (2 * GradientCostForBias) / size
    ]
```

```python
def Gradient(Inputset,Labelset,weight,bias,size):
    dw = 0
    db = 0
    Lrate = 0.001
    for j in range(100000):
        dw , db =  Gradient_Cost_function(Inputset,Labelset,size,weight,bias)
        weight = weight - Lrate*dw
        bias =  bias - Lrate*db
    return [weight,bias] 
```
